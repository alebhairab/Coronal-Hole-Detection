# **Data Visualization (10–20 September 2025, OMNIWeb)**


This plot covers the period **10–20 September 2025**.  
The data was downloaded from [OMNIWeb (NASA GSFC)](https://omniweb.gsfc.nasa.gov/form/omni_min.html).

- **datetime**  
  Used for manipulating dates and times.

- **numpy (as np)**  
  Used for numerical operations, specifically for handling NaN (Not a Number) values during data cleaning.

- **pandas (as pd)**  
  The primary library used for data manipulation, such as reading Excel files and managing the 14-column dataset.

- **matplotlib.pyplot (as plt)**  
  Used for creating the 4-panel multi-plot visualization.

- **matplotlib.dates (as dates)**  
  Specifically used for formatting the x-axis time labels on the final plot.


```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
```

### **Reading Excel Data with Pandas**

This code demonstrates how to load an Excel file into a pandas DataFrame for analysis.

```python
file_path = "/home/bhairab/Documents/isuite/insuite.xlsx"

data = pd.read_excel(file_path)
```

```python
data
```

#### **Replacing Bad Values with NaN**

This code replaces specific placeholder values in the dataset with `NaN` (Not a Number) to handle missing or invalid data 

```python
bad_values = [99999.9,9999.99,9999999, 999.99, 999999, 99.99]
data = data.replace(bad_values, np.nan)
data
```

#### **Adding a Date-Time Column**

This code creates a new column in the DataFrame called `det` that contains a continuous time 

```python
data['det'] = pd.date_range(start='2025-09-10', periods=len(data), freq='5min')
```

```python
data
```

#### **Beautifying Matplotlib Plots**

This code customizes the default appearance of plots in **Matplotlib** by modifying `rcParams`.  
It ensures consistent styling across all figures, making them cleaner 

```python
#beautify plots

import matplotlib as mpl

mpl.rcParams['font.family'      ]='sans-serif'#'Gill Sans'
mpl.rcParams['mathtext.default' ]='regular'
mpl.rcParams['font.size'        ]=12
mpl.rcParams['figure.figsize'   ]=(6,4)
mpl.rcParams['figure.dpi'       ]=180
mpl.rcParams['axes.linewidth'   ]=1.2
mpl.rcParams['xtick.major.size' ]=5
mpl.rcParams['xtick.minor.size' ]=4
mpl.rcParams['ytick.major.size' ]=5
mpl.rcParams['ytick.minor.size' ]=3
mpl.rcParams['xtick.major.width']=1
mpl.rcParams['ytick.major.width']=1
mpl.rcParams['xtick.direction'  ]='out'
mpl.rcParams['ytick.direction'  ]='out'

mpl.rcParams['savefig.bbox'     ]='tight'

mpl.rcParams['animation.html'   ]='jshtml'
```

#### **4-Panel Time Series Visualization (Sept 2025)**

This script generates a compact 4-panel figure using **Matplotlib** to visualize solar wind and geomagnetic parameters from **10–20 September 2025**. The figure is styled with smaller fonts and thinner lines for clarity in 

```python

# --- STYLE SETTINGS ---
mpl.rcParams.update({
    "font.size": 5,
    "axes.titlesize": 5,
    "axes.labelsize": 5,
    "xtick.labelsize": 4,
    "ytick.labelsize": 4,
    "legend.fontsize": 4,
    "lines.linewidth": 0.6,
})

# --- FIGURE SETUP ---
fig, axs = plt.subplots(4, 1, figsize=(4, 2.9), sharex=True, layout="constrained")

# =======================
# PANEL 1: Magnetic Field
# =======================
axs[0].plot(data['det'], data['bx'], label='Bx', color='tab:blue', linewidth=0.8)
axs[0].plot(data['det'], data['by'], label='By', color='tab:orange', linewidth=0.8)
axs[0].plot(data['det'], data['bz'], label='Bz', color='tab:green', linewidth=0.8)
axs[0].set_ylabel('$B_{GSE}$ [nT]', labelpad=2)

ax0_r = axs[0].twinx()
ax0_r.plot(data['det'], data['bnt'], label='|B|', color='black', linewidth=1)
ax0_r.set_ylabel('|B| [nT]', labelpad=2)

# Combine legends
lines_l, labels_l = axs[0].get_legend_handles_labels()
lines_r, labels_r = ax0_r.get_legend_handles_labels()
axs[0].legend(lines_l + lines_r, labels_l + labels_r,
              loc='upper left', ncol=4, frameon=False, handlelength=1)

# =======================
# PANEL 2: Velocity & Temp
# =======================
axs[1].plot(data['det'], data['speed'], label='V', color='tab:orange', linewidth=0.8)
axs[1].set_ylabel('V [km/s]', labelpad=2)

ax1_r = axs[1].twinx()
ax1_r.plot(data['det'], data['tmpr'], label='Tp', color='tab:green', linewidth=0.8)
ax1_r.set_ylabel('Tp [K]', labelpad=2)
ax1_r.set_yscale('log')

# Combine legends
lines_l, labels_l = axs[1].get_legend_handles_labels()
lines_r, labels_r = ax1_r.get_legend_handles_labels()
axs[1].legend(lines_l + lines_r, labels_l + labels_r,
              loc='upper left', ncol=2, frameon=False)

# =======================
# PANEL 3: Density & Pressure
# =======================
axs[2].plot(data['det'], data['density'], label='Np $[cm^{-3}]$', color='tab:red', linewidth=0.8)
axs[2].set_ylabel('$Np [cm^{-3}]$', labelpad=2)

ax2_r = axs[2].twinx()
ax2_r.plot(data['det'], data['pres'], label='P [nPa]', color='tab:purple', linewidth=0.8)
ax2_r.set_ylabel('P [nPa]', labelpad=2)

# Combine legends
lines_l, labels_l = axs[2].get_legend_handles_labels()
lines_r, labels_r = ax2_r.get_legend_handles_labels()
axs[2].legend(lines_l + lines_r, labels_l + labels_r,
              loc='upper left', ncol=2, frameon=False)

# =======================
# PANEL 4: SYM-H
# =======================
axs[3].plot(data['det'], data['symh'], label='SYM-H', color='black', linewidth=1)
axs[3].set_ylabel('SYM-H', labelpad=2)

# Optional legend (if needed)
axs[3].legend(loc='upper left', frameon=False)

# =======================
# FORMATTING
# =======================
axs[3].xaxis.set_major_formatter(dates.DateFormatter('%d'))
axs[3].xaxis.set_major_locator(dates.DayLocator(interval=1))
axs[3].set_xlabel('Day (Sept 2025)')

# Vertical event lines
event_dates = ['2025-09-15 00:00:00', '2025-09-20 00:00:00']
for ax in axs:
    for event in event_dates:
        ax.axvline(pd.Timestamp(event), color='red', linestyle='--',
                   linewidth=0.8,)

# =======================
# SAVE
# =======================
output_path = "/home/bhairab/Documents/isuite/insuiteplot.png"
plt.savefig(output_path, format='pdf', dpi=600)

plt.show()
```
